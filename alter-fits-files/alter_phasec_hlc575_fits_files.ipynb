# Import the necessary packages and define the path to the phaseb_data directory as well as the new directory you would like to save the altered files to. 

```python
from astropy.io import fits
import astropy.units as u
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'image.origin': 'lower'})
import os
from pathlib import Path
import misc

# hlcdir_old = Path('C:/Users/Kian/Desktop/phasec_data/hlc_20190210b'); print(hlcdir_old)
# hlcdir = Path('C:/Users/Kian/Desktop/roman-cgi-phasec-data/hlc'); print(hlcdir)
# hlcdir_old = Path('/Users/kianmilani/Desktop/phasec_data/hlc_20190210b'); print(hlcdir_old)
# hlcdir = Path('/Users/kianmilani/Desktop/roman-cgi-phasec-data/hlc'); print(hlcdir)
hlcdir_old = Path('/groups/douglase/kians-data-files/phasec_data/hlc_20190210b'); print(hlcdir_old)
hlcdir = Path('/groups/douglase/kians-data-files/roman-cgi-phasec-data/hlc'); print(hlcdir)

pupil_diam = 2.363114*u.m
pupil_diam_pix = 309*u.pixel
```

# Add pupil scales to the pupil files.

```python
pupil_fnames = ['pupil.fits', 'pupil_rotated.fits']

for fname in pupil_fnames:
    hlc575pup = fits.open(hlcdir_old/fname)
    hlc575pup_data = hlc575pup[0].data
    hlc575pup.close()
    
    print(hlc575pup_data.shape)
    arr = np.zeros(shape=(312,312))
    arr[1:,1:] = hlc575pup_data
    hlc575pup_data = arr
    
    hdr = fits.Header()
    hdr['PUPLSCAL'] = pupil_diam.value/pupil_diam_pix.value
    hdr['PIXELSCL'] = pupil_diam.value/pupil_diam_pix.value
    hlc575pup_new = fits.PrimaryHDU(data=hlc575pup_data, header=hdr)
    hlc575pup_new.writeto(hlcdir/fname, overwrite=True)
    print(hlcdir/fname)
    print(pupil_diam/pupil_diam_pix, hlc575pup_data.shape)
    misc.myimshow(hlc575pup_data, pxscl=pupil_diam/pupil_diam_pix)
```

# Add pixel scales to the FPM files.

```python
from scipy.ndimage import rotate

def rotate_180(arr):
    M = arr.shape[0]
    N = arr.shape[1]
    rot_arr = np.zeros(shape=(N,M))
    for i in range(M):
        for j in range(N):
            rot_arr[i, N-1-j] = arr[M-1-i, j]
    return rot_arr
            
f = open( str(hlcdir_old/'fpm_files.txt') )
fpm_nlam = int(f.readline())
fpm_lams = np.zeros((fpm_nlam),dtype=float)
for j in range(0,fpm_nlam): fpm_lams[j] = float(f.readline())*1e-6
fpm_root_fnames = [j.strip() for j in f.readlines()] 
f.close()

print(fpm_lams)
print(fpm_root_fnames)

for fpm_rootname in fpm_root_fnames:
    fpm_r = fits.getdata(str(hlcdir_old/fpm_rootname)+'real.fits')
    fpm_i = fits.getdata(str(hlcdir_old/fpm_rootname)+'imag.fits')
    fpm_lambda_m = fits.getheader(str(hlcdir_old/fpm_rootname)+'real.fits')['FPMLAM0M']
    fpm_pxscl_lamD = fits.getheader(str(hlcdir_old/fpm_rootname)+'real.fits')['FPMDX']
    fpm_pxscl_as = (fpm_pxscl_lamD*(fpm_lambda_m/pupil_diam.value)*u.radian/u.pix).to(u.arcsec/u.pix)
    print(fpm_lambda_m, fpm_pxscl_lamD, fpm_pxscl_as)
    
#     fpm_r = rotate(fpm_r, 180)
#     fpm_i = rotate(fpm_i, 180)
#     fpm_r = rotate_180(fpm_r)
#     fpm_i = rotate_180(fpm_i)

    arr = np.zeros(shape=(268,268))
    arr[:-1,:-1] = fpm_r
    fpm_r = arr
    
    arr = np.zeros(shape=(268,268))
    arr[:-1,:-1] = fpm_i
    fpm_i = arr
    
#     arr = np.zeros(shape=(269,269))
#     arr[:-2,:-2] = fpm_r
#     fpm_r = arr
    
#     arr = np.zeros(shape=(269,269))
#     arr[:-2,:-2] = fpm_i
#     fpm_i = arr
    
    print(str(hlcdir/fpm_rootname)+'real.fits', fpm_r.shape)
    hdr = fits.Header()
    hdr['PIXELSCL'] = fpm_pxscl_as.value
    hdr['WAVELENC'] = 0.575e-6
    hdr['PUPLDIAM'] = pupil_diam.value
    hdr['PIXSCLLD'] = fpm_pxscl_lamD
    hlc575fpm_r_new = fits.PrimaryHDU(data=fpm_r, header=hdr)
    hlc575fpm_r_new.writeto(str(hlcdir/fpm_rootname)+'real.fits', overwrite=True)
    
    hdr = fits.Header(); 
    hdr['PIXELSCL'] = fpm_pxscl_as.value
    hdr['WAVELENC'] = 0.575e-6
    hdr['PUPLDIAM'] = pupil_diam.value
    hdr['PIXSCLLD'] = fpm_pxscl_lamD
    hlc575fpm_i_new = fits.PrimaryHDU(data=fpm_i, header=hdr)
    hlc575fpm_i_new.writeto(str(hlcdir/fpm_rootname)+'imag.fits', overwrite=True)
    
    if '0.57500000um' in fpm_rootname or '0.54625000um' in fpm_rootname or '0.60375000um' in fpm_rootname:
        misc.myimshow2(fpm_r, fpm_i, pxscl=fpm_pxscl_as)
    
    
    
    
```
